# Use Pylab to Graph Comuptational Complexity

We will use pylab to graph out computational complexity in in our code.

```python
import pylab as plt
```

## Retirement Growth Rates

In the following section, we will use pylab to visualize how retirement increases in proportion to investments. First we will calculate the growth rate  of retirment savings amount based on monthly payment, annual % increase, and  number of terms. Then we will implement a helper functions:

- `displayRetirementWithMonthlies()` which tracks retirement payout with various monthly payments and a static growth rate of 5%. 

- `displayRetirementWithRates()` which tracks retirement paytou given static payment and various growth rates.

- `displayRetirementWithMonthsAndRates()` which tracks retirement paytou given various payment amt and various growth rates.


#### Calculate Growth Rate

This function takes in `monthly` a monthly contribution to retirement, `rate` the yearly growth rate, and `terms` the number of savings terms where money was contributed to retirement.

`savings` (an empty list to begin) will be filled with total savings as terms go on.
`base` is the term number
`monthlyRate` is yearly rate / 12 months

The for loop iterates through every base and:
- appends term `i` to list `base`
- appends the (interest accrued from savings + current month's contribution)

The lists `base`, `savings` are returned.

```python
def retire(monthly, rate, terms):
    savings = [0]
    base = [0]
    monthlyRate = rate/12
    for i in range(terms):
        base += [i]
        savings += [savings[-1]*(1 + monthlyRate) + monthly]
    return base, savings
```

The function `displayRetirementWithMonthlies` will run monthly retirements with various savings rates and graph the end difference in retirement.

- We define a figure.
- Clear the frame.
- Iterate through various monthly contribution sizes and return projected savings over `terms`.
    - Plot savings growth rate for various contribution amounts to a pylab chart
    - generate a legend for the plots of each contribution amount

#### Plot Various Monthly Payment

Okay, money invested for retirement isn't exactly a payment, but it looked nicest in the title. This  helper function uses pylab to plot out how retirement money changes as we increase monthly savings, from 500 to 1100.

```python
def displayRetirementWithMonthlies(monthlies, rate, terms):
    plt.figure('retireMonth')
    plt.clf()
    for monthly in monthlies:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, label = 'retire:'+str(monthly))
        plt.legend(loc = 'upper left')
```

```python
displayRetirementWithMonthlies([500, 600, 700, 800, 900, 1000, 1100], .05, 40* 12)
```

#### Plot Various Growth Rates

The following helper function is a remix of the above helper function. It runs `retire()` with various growth rates.

```python
def displayRetirementWithRates(monthly, rates, terms):
    plt.figure('retireRate')
    plt.clf()
    for rate in rates:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, 
                 label = 'retire:'+str(monthly)+ ':' + str(int(rate*100)))
        plt.legend(loc = 'upper left')
```

```python
displayRetirementWithRates(800,[.03, .05, .07], 40*12)
```

#### Plot Various Growth Rates and Payments

This last helper function plots changes in payment amoun annd growth rate. In the end there are will be 12 lines plotted out.

```python
def displayRetirementWithMonthsAndRates(monthlies, rates, terms):
    plt.figure('retireBoth')
    plt.clf()
    plt.xlim(30*12, 40*12)
    for monthly in monthlies:
        for rate in rates:
            xvals, yvals = retire(monthly, rate, terms)
            plt.plot(xvals, yvals,
                     label = 'retire:'+str(monthly)+ ':' + str(int(rate*100)))
            plt.legend(loc = 'upper left')
        
```

```python
displayRetirementWithMonthsAndRates([500, 700, 900, 1100], 
                                       [.03, .05, .07], 40*12)
```

Notice this gets super messy with 12 lines. Even if they were ll labeled, it's just hard to sort through each line and find the right color to compare with the legend. Let's modify the above function to give different line types for each payment type. As we iterate through, the func uses modulo to check what the monthly savings size and given growth rate are.

```python
def displayRetirementWithMonthsAndRates(monthlies, rates, terms):
    plt.figure('retireBoth')
    plt.clf()
    plt.xlim(30*12, 40*12)
    monthLabels = ['r', 'b', 'g', 'k']
    rateLabels = ['-', 'o', '--', '^']
    for i in range(len(monthlies)):
        monthly = monthlies[i]
        monthLabel = monthLabels[i%len(monthLabels)]
        for j in range(len(rates)):
            rate = rates[j]
            rateLabel = rateLabels[j%len(rateLabels)]
            xvals, yvals = retire(monthly, rate, terms)
            plt.plot(xvals, yvals, monthLabel+rateLabel,
                    label = 'retire:'+str(monthly)+ ':' + str(int(rate*100)))
            plt.legend(loc = 'upper left')  
```

```python
displayRetirementWithMonthsAndRates([500, 700, 900, 1100], [.03, .05, .07], 40*12)
```

Of course, if we had more elements in `monthlies` or `rates` than in `monthLabels` or `rateLabels`, the modulo would repeat colors/line-styles. But there would still be a visual breakup.
